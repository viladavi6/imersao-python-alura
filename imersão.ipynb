```python
df_tot = pd.read_excel('C:/Users/DVILA/Documents/Imersão Python - Alura/desafio.xlsx', sheet_name='Total_de_acoes')
df_tot
```

```python
df_ticker = pd.read_excel('C:/Users/DVILA/Documents/Imersão Python - Alura/desafio.xlsx', sheet_name='Ticker')
df_ticker
```

```python
df_gpt = pd.read_excel('C:/Users/DVILA/Documents/Imersão Python - Alura/desafio.xlsx', sheet_name='Chatgpt')
df_gpt
```

```python
df_pri = df_pri[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_pri
```

```python
df_pri = df_pri.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'Var_dia'}).copy()
df_pri
```

```python
df_pri['Var_pct'] = df_pri['Var_dia'] / 100
df_pri['valor_inicial'] = df_pri['valor_final'] / (df_pri['Var_pct'] + 1)
df_pri
```

```python
df_pri = df_pri.merge(df_tot, left_on='Ativo', right_on='Código', how='left')
df_pri
```

```python
df_pri = df_pri.drop(columns=['Código'])
df_pri
```

```python
df_pri['Variacao_rs'] = (df_pri['valor_final'] - df_pri['valor_inicial'])* df_pri['Qtde. Teórica']
df_pri
```

```python
df_pri['Qtde. Teórica'] = df_pri['Qtde. Teórica'].astype(int)
df_pri
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_pri
```

```python
df_pri = df_pri.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_pri
```

```python
df_pri['resultado'] = df_pri['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_pri
```

```python
df_pri = df_pri.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_pri = df_pri.drop(columns=['Ticker'])
df_pri
```

```python
df_gpt
```

```python
df_pri = df_pri.merge(df_gpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_pri = df_pri.drop(columns=['Nome da empresa'])
df_pri
```

```python
df_pri['faixa_idade'] = df_pri['Idade (anos)'].apply(lambda x: 'Mais de 100' if x > 100 else('Menos de 50' if x < 50 else 'Entre 50 e 100' ))
df_pri
```

```python
df_pri = df_pri.rename(columns={'Idade (anos)':'idade'}).copy()
df_pri
```

```python
maior = df_pri['Variacao_rs'].max()
menor = df_pri['Variacao_rs'].min()
media = df_pri['Variacao_rs'].mean()
media_subiu = df_pri[df_pri['resultado'] == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_pri[df_pri['resultado'] == 'Desceu']['Variacao_rs'].mean()

print(f'Maior:\tR$ {maior:,.2f}')
print(f'Menor:\tR$ {menor:,.2f}')
print(f'Média:\tR$ {media:,.2f}')
print(f'Média Subiu:\tR$ {media_subiu:,.2f}')
print(f'Média Desceu:\tR$ {media_desceu:,.2f}')
```

```python
df_pri_subiu = df_pri[df_pri['resultado'] == 'Subiu']
df_pri_subiu
```

```python
df_analise_segmento = df_pri_subiu.groupby('Segmento') ['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_pri.groupby('resultado') ['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Resultado')
fig.show()
```

```python
pie = px.pie(df_analise_segmento, values='Variacao_rs', title='Variação em Reais dos Segmentos')
pie.show()
```
